Escreva um programa que faça a validação de um CPF informado pelo usuário (veja a regra de validação no arquivo "Validação CPF" abaixo) e informe ao final se o CPF é válido ou não.

Os CPF's com todos os números iguais (11111111111, 22222222222, ..., 99999999999) devem ser considerados inválidos, apesar de serem considerados válidos pela regra de validação do CPF.

Ao final, o programa deve apresentar uma das seguintes mensagens: "CPF válido!", "CPF inválido!" ou "CPF inválido (números iguais)!"

```python
mult = []

cpf = input('Digite seu cpf: ')
if cpf.count(cpf[0]) != 11 and len(cpf) == 11: 
  # PRIMEIRO DIGITO VERIFICADOR - X1
  cpf9 = list(cpf[:-2])
  mult = [valid * int(num) for valid, num in zip(reversed(range(2, 11, 1)) , cpf9)]
  result = 11 - (sum(mult) % 11) 
  x1 = 0 if result > 9 else result

  # SEGUNDO DIGITO VERIFICADOR - X2
  if  int(cpf[9]) == x1:
    cpf10 = list(cpf[:-1])
    mult = [valid * int(num) for valid, num in zip(reversed(range(2, 12, 1)), cpf10)]
    result = 11 - (sum(mult) % 11) 
    x2 = 0 if result > 9 else result

  if int(cpf[10]) == x2: 
    print("CPF {} válido!".format(cpf))
  else:
      print("CPF inválido! ")
else:
  print("CPF inválido! Deve conter 11 números diferentes.")
```

Escreva um programa que a partir da entrada dos 9 primeiros dígitos de um CPF, calcule o seu primeiro e o seu segundo dígito verificador. Em seguida, apresente o CPF completo.

```python
mult = []

cpf = input('Digite os 9 primeiros dígitos do seu cpf: ')
if cpf.count(cpf[0]) != 9 and len(cpf) == 9: 
  # PRIMEIRO DIGITO VERIFICADOR - X1
  mult = [valid * int(num) for valid, num in zip(reversed(range(2, 11, 1)) , list(cpf))]
  result = 11 - (sum(mult) % 11) 
  x1 = 0 if result > 9 else result

  # SEGUNDO DIGITO VERIFICADOR - X2
  mult = [valid * int(num) for valid, num in zip(reversed(range(2, 12, 1)), list(cpf + str(x1)))]
  result = 11 - (sum(mult) % 11) 
  x2 = 0 if result > 9 else result

  print("CPF: {}.{}.{}-{}{}".format(cpf[0:3], cpf[3:6], cpf[6:9],x1,x2))

else:
  print("CPF inválido! Deve conter 11 números diferentes.")
```

Atividade 3 

Escreva um programa em Python para ler uma sequencia de números inteiros (a sequencia termina quando o valor zero for digitado) e armazenar os valores lidos em uma lista. Em seguida, o programa deve particionar os elementos dessa lista pelo seu valor médio, gerando duas outras listas: a primeira lista contendo os elementos que tenham valores menores que a média e a segunda lista contendo os elementos com valores maiores ou iguais à média. Observação: utilize list comprehension para o preenchimento da primeira e da segunda lista.

```python
numbers = []
while 1:
  val = int(input('Digite um número: '))
  if val == 0: break
  numbers.append(int(val))
  
media = round(sum(numbers) / len(numbers),1)
numbers.sort() 
middle = [ x for x in numbers if x >= media ]
Abaixo_media = numbers[ : numbers.index(middle[0])]
Acima_media = numbers [ numbers.index(middle[0]) : ]

print('Lista abaixo da média: ', Abaixo_media)
print('Lista acima da média: ', Acima_media)
```

Atividade 4

Escreva um programa em Python que leia 5 números inteiros positivos e armazene os valores lidos em uma lista. Em seguida escreva "Sim!" os números foram digitados em ordem crescente e "Não!" caso contrário.

```python
import functools 

numbers = [int(input('Digite um número: ')) for x in range(5)]
numbers_validation =  numbers.sort() 

if functools.reduce(lambda x, y : x and y, map(lambda p, q: p == q, [numbers] , [numbers_validation ]), True) or all(t==numbers[0] for t in numbers): 
    print ("Sim!") 
else: 
    print ("Não!") 
```
